## 확률
표본공간(S)

- 사건이 일어날 확률

- 고전적 확률 정의
      
      실험(나올 수 있는))결과가 N 가지 일 때 일어난 사건A가 m번 일어났으면 P(A) = m/N
      각 실험 가능성은 모두 같을 때
      
      (예시 주사위 {1,2,3,4,5,6} 1/6 1/6 1/6 1/6)

- 경험적 정의
      반복 횟수를 N이라 했을 때 무한번 반복했을 때 발생한는 A 횟수 m
$P(A) = lim \frac{m}{N}$

- 공리적 정의
      콜모코로프 공리적 정의 세가지(아래 조건 3개를 모두 만족해야함!!)
      
      1. 0 <= P(A) <= 1 

      2. P(S) = 1
      
      3. 실험이 표본공간의 부분집합 이면서 다른 사건과 교집합이 없을 때 각 실험의 확율을 더한 값과 합집합의 확률과 같다.
      
      예를 들면 동전 두개 던지기를 합니다.
      앞면 : H, 뒷면, T
      표본공간(나올 수 있는 모든 경우) = S{HH, HT, TT, TH}
      앞이 안 나오는 사건(A{TT}) 과 뒷면이 안 나올 사건(B{HH})면 A와 B는 교집합이 없습니다. (A와 B는 배반사건-공통된 부분이 없음)
      P(AUB) = P({TT,HH}) = P({TT}) + P({HH})
      P(AUB) = P(A) + P(B)


## 확율의 성질
- 공집합의 확율 0
      
      예) 주사위를 던져서 아무것도 안나올 확률 - 0%


- 여집합의 확률

      실험 A의 여집합은 1-P(A)

- 포함-배제 원리

      P(AUB) = P(A) + P(B) - P(AnB) (공통된 것이 나올 확률이 있는경우, 그 경우를 다시 빼줌)

- 각 사건이 배반사상(event)일 경우

      P(A1 U A2 U A3 .....U An) = P(A1)+P(A2)....P(An)
      A가 B의 부분집합일 때 P(A) <= P(B)

## 조건부 확률
- 어떤 사건 B가 일어날 때(P(B) > 0) 사건 A가 일어날 확률

    예) A: 과일에 벌레가 있음 / B: 과수원에서 사과를 땀
    
    과수원에서 사과를 땄을 때, 그 사과에 벌레가 있을 확률
        

$$P(A|B) = \frac{P(A\cap B)}{P(B)}$$

- 사건 A가 발생했을 때 사건 B가 발생할 확률

    위의 예와 비슷
$$P(B|A) = \frac{P(A\cap B)}{P(A)}$$

$$P(A\cap B) = P(A|B)P(B) = P(B|A)P(A)$$

- 독립사건 (교집합이 없는 경우, A가 일어났을때 B가 일어날 확률)

$$P(A|B) = P(A)$$
$$P(B|A) = B(B)$$

## 베이즈 정리

- 분할사건
      
      사건 K개가 일어났을 때 A1 ... AK 서로 다 교집합이 없는경우 (분할 사건, 각각 배반사건)
      
      사건 K개를 모두 합집합했을 때 S가 되는 경우
      
      예) A:과수원에 과일(사과, 딸기, 포도 ...)밭이 겹치지 않게 각각 심어져 있음

          B:과수원에 불이 나서 과일 밭 과일들이 부분부분이 타들어감
          
          이 경우 타버린 과일들(B)의 총량은 각각 타버린(B) 과일들(A)의 계수의 총합

$$ B = (B \cap {A_1}) U ⋅⋅⋅ U (B \cap {A_k}) $$



$$ P(B) = P(B \cap {A_1}) + ⋅ ⋅ ⋅ + P(B ∩ {A_k}) $$



$$P({A_i}|B) = \frac{P(B|{A_i})P({A_i})}{P(B)} = \frac {P(B|{A_i})P({B_i})}{P(B|{A_1})P({A_1}) + P(B|{A_2})P({A_2})+ ⋅ ⋅ ⋅+ P(B|{A_k})P({A_k})}$$

사전 확률을 사후확률로 추정하는 정리

## 확률 변수와 확률 분포

- 확률변수

      표본공간에서 정의된 실수값 함수
      확률 변수들의 통계량 찾고, 확률 변수를 통해서 모수를 추정
      대문자 X 로 표현

- 확률 분포

      확률 변수 값과 확률을 대응시키는 표, 그래프, 함수


- 이산확률 변수

      countably infinite 변수
      확률 질량 함수를 통해 확률을 얻을 수 있다.

  - 확률 질량 함수

      P(a <= X <= b) =  𝚺f(x)
      


- 연속확률 변수

      하나로 특정할 수 없고 셀수 없는 변수

    - 확률 밀도 함수
          
          확률 밀도 함수 f(x)는 연속확률 변수 X에대한 확률에 분포를 나타내는 함수
          모든 x에 대해서 f(x) >= 0
          확률 밀도 함수의 전체 면적은 1
          특정 구간에서의 확률은 확률밀도 함수를 적분한 값이다.

## 정규 분포
- 정의

      확률 변수 X가 평균이 m이고 분산이 r인 정규분포는 X ~ N(m, r)

- 특징

      모든 정규분포는 좌우 대칭
      정규곡선의 위치는 평균에 의해서 결정된다.
      표본공간에 따라 평균과 표준편차는 다르지만 평균으로부터 k배 표준편차 이내에 확률 변수 X가 존재할 확율을 동일




## 표준 정규 분포
- 확률 변수를 정규화를 시켜서 표준편차 1이고 평균이 0인 정규분포로 만든 것

## 기댓값
- 모집단의 분포의 모양을 표현할 수 있는 특성값

- 이론적으로 확률 변수의 확률밀도 함수를 알면 평균값 확률 변수의 기댓값

- 이산확률 변수의 기댓값

$$E[X] = Σ{x_i^2}p({x_i})$$


- 기대값의 성질
    
    모집단(알 수는 있지만 값으로 활용 불가)의 평균 = 기댓값


## 분산과 표준편차

- 분산 확률분포함수의 모양을 설명할 수 있다.
- 분산은 확률분포함수에 확률이 모여있는지 퍼져있는지를 알 수 있는 수치적 파라미터

$${σ^2} = Var[X] = E[{(X-\mu)^2}]$$

- 성질
      Var[X] >= 0
      Var[cX] = c^2Var[X]

      Var[X+Y] = Var[X] + Var[Y] + 2E[(X-m_x)(Y-m_y)]



- 표본평균의 분산과 표본의 분산의 관계
$$E[X] = E[\bar{X}]$$
$$Var[\bar{X}] = \frac{1}{N}Var[X]$$
