```python
bus_st = bus.groupby(['노선번호','역명']).mean().reset_index()
```

```python
bus_st = bus_st.drop(columns=['사용일자'])
bus_st
```

```python
bus_get_on = pd.DataFrame()
bus_get_off = pd.DataFrame()

#승차 데이터 프레임
bus_get_on['역명'] = bus_st['역명']
bus_get_on['노선번호'] = bus_st['노선번호']

# 하차 데이터프레임
bus_get_off['역명'] = bus_st['역명']
bus_get_off['노선번호'] = bus_st['노선번호']

for column in bus_st.columns:
  if '승차총승객수' in column:
    bus_get_on[[column]] = bus_st[[column]]
  elif '하차총승객수' in column:
    bus_get_off[[column]] = bus_st[[column]]

bus_get_on = bus_get_on.set_index('역명')
bus_get_off = bus_get_off.set_index('역명')

bus_get_on

```

```python
df = pd.DataFrame(index=bus_st['역명'])
df['노선번호'] = bus_get_on['노선번호']
df['평균승차인원수'] = bus_get_on[bus_get_on.columns.difference(['노선번호'])].mean(axis=1)
df['평균하차인원수'] = bus_get_off[bus_get_off.columns.difference(['노선번호'])].mean(axis=1)
df['평균승하차인원수'] = df[['평균승차인원수','평균하차인원수']].sum(axis=1)
df.loc['종로2가사거리(00067)']
```

```python
df.sort_values('평균승하차인원수', ascending=False)[['평균승차인원수','평균하차인원수']].head(10).plot(kind=('bar'))
plt.show()
```

```python
# !pip install geopy
from geopy.geocoders import Nominatim

def geocoding(address):
  geolocoder = Nominatim(user_agent='South Korea',  timeout=None)
  geo = geolocoder.geocode(address)
  if geo == None:
    return None
  crd = [str(geo.latitude), str(geo.longitude)]
  return crd

geocoding('종로2가사거리(00067)')
```
