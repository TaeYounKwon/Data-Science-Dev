# Data Visualization

데이터 품질 및 분포 특성을 시각적으로 확인하기 위해 사용


데이터 이해 및 분석 결과를 설들력 있게 표현하기 위해 사용


모집단 : 분석이 대상이 되는 전체


표본집단 : 모집단에서 추출된 모집단의 부분집합

모집단의 특성을 추정하기 위해서 통계적인 기법을 사용

```python
# Check with Setosa
ax = iris[iris['species'] == 'setosa']['petal_length'].plot.hist(density=True)
iris[iris['species'] == 'setosa']['petal_length'].plot.density(ax=ax)
plt.show()

# From the graph,
# Highest value = Mode
```

**왜도(Skewness)**

데이터의 분포에 따라서 오른쪽으로 치우쳤는지 왼쪽으로 치우쳤는지를 확인확인

왜도가 중앙으로 eqaul하게 퍼져있으면 (Mean = Median = Mode)

Nagative skewness : 데이터의 중심이 오른쪽이 치우쳐졌을때 (Mean < Median < Mode)

Positive skewness : 데이터의 중심이 왼쪽에 치우쳐졌을때 (Mode < Median < Mean)



**첨도(Kurtosis)**


데이터가 얼마나 평균에 모여있는지 (얼마나 뾰족한지!)

Leptokurtic(고첨) : 정규분포보다 많이 몰려 있어서 뾰족


Mesokutic(중첨) : 정규 분포와 유사


Pltykutic(저첨) : 정규 분포보다 완만

```python
fig, ax = plt.subplots(1,1)

for species in iris['species'].unique(): # .unique remove the duplicates, and brings the only unique values
  width_series = iris[iris['species']==species]['sepal_width']
  width_series.plot.hist(alpha = 0.5, ax=ax, label=species)
plt.legend()
plt.show()
```

```python
# Box Plot
iris.boxplot(by='species', layout=(2,2), figsize=(12,8))
plt.show()
```

```python
# Violin plot
fig, ax = plt.subplots(2,2, figsize=(12,8))
for i, col in enumerate(['petal_length','petal_width','sepal_length','sepal_width']):
  sns.violinplot(data=iris, x='species', y=col, ax=ax[i//2][i%2]) # sns - seaborn, automatically coloring the violin plot

plt.show()                 
```

## 단어
**모집단**
* 조사 대상이 되는 모든집단

**표본(Sampling)집단**
* 모집단의 부분집합

**모수**
* 모집단의 어떤 값

**통계량**
* 표본 집단의 정보


## 데이터의 중심

**최빈값(mode)**

* 빈도가 가장 높은값
* 범주형 데이터의 대표값
* 2개 이상 존재 가능


**중앙값**
* 데이터를 내림차순 해서 가장 중간에 있는 데이터
* 짝수개 일 경우는 중앙에 있는 데이터 두개를 더하고 나눈 값이 중앙값


**산술평균**
* 일반적으로 사용하는 평균
* 극단값 이상치에 영향을 많이 받는 통계량
* 데이터의 개수로 나눔
* 절삭평균으로 극단값에 대해 보정

**가중평균**
* 일반적인 산술 평균 각각의 데이터마다 중요도에 따라서 가중치
* 가중치를 다 더한 값으로 나눠줌

**기하평균**
* 성장률이나 증가율 이런 비율에 대한 데이터를 평균 낼 때 사용
* 음수나 0의 데이터가 있는 경우에는 적용을 못합니다.

**편차**
* 데이터 값과 평균을 빼준 값

**분산**
* 편차의 제곱의 합을 데이터 수로 나눈 값
* 편차를 다 더하면 0나오기 때문에 제곱을 진행
* 편차가 큰 데이터 값에 대해 영향정도를 키우기 위해서

**표준편차**
* 분산의 제곱근을 한 값

### 절삭평균
산술평균 사용시 IQR x 1.5배 되는 값이 극단적 영향을 데이터의 분포에 줄 때 사용

IQR x 1.5배 되는 데이터들(이상치)들을 다 버리고 평균을 구함

더 평균에 밀집된 데이터를 볼 수 있음

```python
from scipy.stats import trim_mean
data = np.array([1, 2, 2, 3, 3, 3, 4, 4, 5, 10])
mean_ = np.mean(data)
trim_mean_ = trim_mean(data, 0.1) # trim_mean(data, how many percentage(0.1 = 10%))
median = np.median(data)
print(f'''
mean = {mean_}
절삭평균 = {trim_mean_}
중앙값 = {median}
''')
```

```python
from scipy.stats import trim_mean
data = np.array([1, 2, 2, 3, 3, 3, 4, 4, 5, 10, 9999999999])
mean_ = np.mean(data)
trim_mean_ = trim_mean(data, 0.1) 
median = np.median(data)
print(f'''
mean = {mean_}
절삭평균 = {trim_mean_}
중앙값 = {median}
''')
```

```python
data = np.array([1, 2, 2, 3, 3, 3, 4, 4, 5, 10, 9999999999])
data_length = len(data)

if data_length %2 == 0:
  idx = data_length//2
  median = (data[idx] + data[idx+1])/2

else:
  idx = data_length//2
  median = data[idx]
  
print('중앙값 : ', median)
```

## 편차 표준편차 분산

```python
norm = np.random.normal(10.0, 2.0, 100) # np.random.normal(center value, +/- value, number amount)
```

```python
# 평균
norm_mean = norm.mean()
norm_mean
```

```python
# 편차
norm_dev = [i-norm_mean for i in norm]
```

```python
# 분산
norm_var = sum([i**2 for i in norm_dev])/(len(norm) - 1) #모집단 -1로 나눠져야 모수를 더 잘 표현할 수 있다(?)
norm_var
```

```python
# 표준편차
import math
math.sqrt(norm_var)
# np.std(norm_var) # 넘파이 내 표준편차 구하는 메서드
```
