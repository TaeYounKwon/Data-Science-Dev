```python
import pandas as pd
import numpy as np
```

```python
df1 = pd.DataFrame({
    'date' : [1101,1102,1103,1104,1105],
    'temperature' : [20,18,16,14,5]
    
}, columns=['date','temperature'])
df1
```

```python
df2 = pd.DataFrame(data={
    'date' : [1101,1102,1103,1104,1109,1120],
    'humidity': [10,20,40,100,0,50]
}, columns=['date','humidity'])
df2
```

```python
# Combine df1 and df2
# only merge if there is same number of index
pd.merge(df1,df2) 
```

```python
# Merge and bring all data even though it's not equal number of index
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='left') # 약간 sql 문법이랑 비슷한듯 ㅇㅇ
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='right',on='date')
```

```python
df1 = pd.DataFrame({
    'date' : [1101,1102,1103,1104,1105],
    'temperature' : [20,18,16,14,5]
    
}, columns=['date','temperature'])
df1
```

```python
df2 = pd.DataFrame({
    'code' : [1101,1102,1103,1104,1105],
    'humidity' : [20,18,16,14,5]
    
}, columns=['code','humidity'])
df2
```

```python
pd.merge(df1,df2,left_on='date',right_on='code')
```
