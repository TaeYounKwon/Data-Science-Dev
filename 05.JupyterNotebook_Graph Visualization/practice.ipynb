### Bar Graph

```python
data = {
    'banana': 3, 'apple': 2, 'orange':10
}
# line=> plt.plot // bar=> plt.bar or // vertical bar=> plt.barh
print(data.keys(),data.values())
plt.bar(data.keys(),data.values())
plt.title('Amount of Fruit')
plt.xticks([0,1,2],data.keys())
plt.yticks(sorted(data.values()))
plt.grid(True)
plt.xlabel('Fruit')
plt.ylabel('Amount')
plt.show()
```

```python
# Draw Bar graph vertical ways
data = {
    'banana': 3, 'apple': 2, 'orange':10
}
# line=> plt.plot // bar=> plt.bar
# print(data.keys(),data.values())
plt.barh([0,1,2],data.values())
plt.title('Amount of Fruit')
plt.yticks([0,1,2],data.keys())
plt.xticks(sorted(data.values()))
plt.grid(True)
plt.ylabel('Fruit')
plt.xlabel('Amount')
plt.show()
```

### Pie Chart

```python
ratio = [34, 32, 16, 18]
labels = ['Apple','Banana','Melon','Graphs']
explode =[0.05,0.05,0.05,0.05] # how far away from center
color=['red','gold','green','purple']

# Pie char method => plt.pie
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, shadow=True, explode= explode, colors=color)
plt.show()
```

```python
# draw different version of graph
wedgeprops = {'width':0.7, 'edgecolor':'w','linewidth':5}
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, shadow=True, explode= explode, colors=color, wedgeprops=wedgeprops)
plt.show()
```

```python
# draw perfect circle graph
plt.pie(ratio, labels=labels, autopct='%.1f%%', startangle=260, counterclock=False, colors=color)
plt.show()
```

### Histogram

```python
# Method => plt.hist
# histogram
# with bins, it shows how the data are spreaded with the number of bins
heights = [168, 181, 164, 156, 178, 174, 161, 177, 166, 168, 159, 171, 180, 159, 167, 181, 169, 173, 169, 174, 170, 165]
plt.hist(heights, bins=10, label='heights')
plt.title('heights Data Histogram')
plt.ylabel('Number of People')
plt.xlabel('Heights')
plt.show()
```

```python
plt.hist(heights, bins=10, label='heights', cumulative=True)
plt.title('Cumulative Heights Data Histogram')
plt.ylabel('Number of People')
plt.xlabel('Heights')
plt.show()
```

```python
plt.hist(heights, bins=10, label='heights',histtype='step')
plt.title('Heights Data Step Histogram')
plt.ylabel('Number of People')
plt.xlabel('Heights')
plt.show()
```

```python
plt.hist(heights, bins=10, label='heights',histtype='bar')
plt.title('Heights Data Bar Histogram')
plt.ylabel('Number of People')
plt.xlabel('Heights')
plt.show()
```

```python
plt.hist(heights, bins=10, label='heights', density=True, alpha=0.6, color='red')
plt.title('Heights Data Histogram')
plt.ylabel('Density of People')
plt.xlabel('Heights')
plt.show()

# Adding up all the area in the graph = 1
```

### Box Graph

```python
plt.style.use('default')
data_a = np.random.normal(0,2.0,1000)
data_b = np.random.normal(-3.0,1.5,500)
data_c = np.random.normal(1.2,1.5,1500)

```

```python
fig, ax = plt.subplots()
# Method => plt.boxplot
ax.boxplot([data_a,data_b,data_c], notch=True, whis=1.5, vert=True)
ax.set_ylim(-10.0,10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

```python
fig, ax = plt.subplots()
# Method => plt.boxplot

# Increase the whis value
ax.boxplot([data_a,data_b,data_c], notch=True, whis=3, vert=True)
ax.set_ylim(-10.0,10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

```python
# Change the Data Range
data_a = np.random.normal(0,5.0,1000)
data_b = np.random.normal(-3.0,0.5,500)
data_c = np.random.normal(1.2,3,1500)

fig, ax = plt.subplots()
# Method => plt.boxplot
ax.boxplot([data_a,data_b,data_c], notch=True, whis=1.5, vert=True)
ax.set_ylim(-20.0,20.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')
plt.show()
```

```python
# Horizontal Graph
fig, ax = plt.subplots()
ax.boxplot([data_a,data_b,data_c], notch=True, whis=1.5, vert=False)
# ax.set_ylim(-20.0,20.0)
# ax.set_xlabel('Data Type')
# ax.set_ylabel('Value')
plt.show()
```

### Violin Plot

```python
np.random.seed(0)
data_a = np.random.normal(0,2.0,1000)
data_b = np.random.normal(-3.0,1.5,500)
data_c = np.random.normal(1.2,1.5,1500)
```

```python
violin = plt.violinplot([data_a,data_b,data_c],showmeans=True, 
                        showextrema=True, showmedians=True)

plt.xticks(np.arange(1,4),labels=['A','B','C'])
plt.show()
```

```python
fig, ax = plt.subplots()
# Method => plt.boxplot
ax.boxplot([data_a,data_b,data_c], notch=True, whis=1.5, vert=True)
ax.set_ylim(-10.0,10.0)
ax.set_xlabel('Data Type')
ax.set_ylabel('Value')

violin = plt.violinplot([data_a,data_b,data_c],showmeans=True, 
                        showextrema=True, showmedians=True)

plt.xticks(np.arange(1,4),labels=['A','B','C'])
plt.title('Comparing b/w Box and Violin Plot')
plt.show()
```

```python
violin = plt.violinplot([data_a,data_b,data_c],showmeans=True, 
                        showextrema=True, showmedians=True)

plt.xticks(np.arange(1,4),labels=['A','B','C'])
violin['bodies'][0].set_facecolor('red')
violin['cbars'].set_edgecolor('green')
violin['cmaxes'].set_edgecolor('navy')
violin['cmins'].set_edgecolor('purple')
violin['cmeans'].set_edgecolor('red')
violin['cmedians'].set_edgecolor('orange')
plt.show()
```
